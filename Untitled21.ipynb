```python
import pandas as pd

# Replace 'twitter_training.csv' with the actual name of your uploaded file
file_path = 'twitter_training.csv'
data = pd.read_csv('/content/twitter_training.csv')

# Correct the column headers
column_names = ["id", "topic", "sentiment", "content"]
data.columns = column_names

# Display the first few rows of the dataset
data.head()
```

```python
# Check for missing values
missing_values = data.isnull().sum()
print("Missing values:\n", missing_values)

# Distribution of sentiments
sentiment_distribution = data['sentiment'].value_counts()
print("Sentiment distribution:\n", sentiment_distribution)

# Distribution of topics
topic_distribution = data['topic'].value_counts()
print("Topic distribution:\n", topic_distribution)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set up the plotting environment
sns.set(style="whitegrid")

# Bar plot for sentiment distribution
plt.figure(figsize=(10, 6))
sns.countplot(x='sentiment', data=data, order=data['sentiment'].value_counts().index, palette='viridis')
plt.title('Sentiment Distribution')
plt.xlabel('Sentiment')
plt.ylabel('Count')
plt.show()
```

```python
# Bar plot for topic distribution
plt.figure(figsize=(14, 8))
sns.countplot(y='topic', data=data, order=data['topic'].value_counts().index, palette='viridis')
plt.title('Topic Distribution')
plt.xlabel('Count')
plt.ylabel('Topic')
plt.show()
```
